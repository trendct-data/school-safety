```python
import pandas as pd
```

```python
bullied_at_school = pd.read_csv("data/clean/school safety 2015 - Bullied at school clean.csv")
bullied_online = pd.read_csv("data/clean/school safety 2015 - Bullied online - clean.csv")

bullied_at_school_nat = bullied_at_school[bullied_at_school["State"] == "United States"]
bullied_online_nat = bullied_online[bullied_online["State"] == "United States"]

bullied_at_school = bullied_at_school[bullied_at_school["State"] != "United States"]
bullied_online = bullied_online[bullied_online["State"] != "United States"]
```

### Bullying

Bullying was defined for respondents as "when one or more students tease, threaten, spread rumors about, hit, shove, or hurt another student over and over again." "On school property" was not defined for survey respondents.
